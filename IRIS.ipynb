```python
# Aplicação K-means no data set IRIS
```

```python
import Kmeans
import numpy as np
```

```python
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
model = Kmeans.KMeans(K = 3, plot_steps = True)
```

```python
model.K
```

```python
iris = iris.data[:,:2]
```

```python
model.predict(iris)
```
